```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogState) -> BlogState:
    # fetch title
    title=state["title"]
    
    
    #call llm to create outline
    prompt=f"Create a detailed outline for a blog post with the title: {title}"
    outline=model.invoke(prompt).content
    # return state with outline
    state["outline"]=outline
    return state
```

```python
def create_blog(state: BlogState) -> BlogState:
    # fetch title and outline
    title=state["title"]
    outline=state["outline"]
    
    #call llm to create blog content
    prompt=f"Write a detailed blog post with the title: {title} and the following outline: {outline}"
    content=model.invoke(prompt).content
    
    # return state with content
    state["content"]=content
    return state
```

```python
#Graph
graph=StateGraph(BlogState)

#Nodes
graph.add_node("create_outline",create_outline)
graph.add_node("create_blog",create_blog)

#Edges
graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","create_blog")
graph.add_edge("create_blog",END)

#complie
workflows=graph.compile()
```

```python
intial_state={"title":"The future of AI"}
final_state=workflows.invoke(intial_state)
print(final_state["content"])
```
