```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
    
```

```python
def llm_Qa(state:LLMState)->LLMState:
    response=state["question"]
    
    #Form a prompt 
    prompt=f'Answer the following question: {response}'
    
    #ask that Question to the model
    answer=model.invoke(prompt).content
    
    #Update the state with the answer
    state["answer"]=answer
    
    return state
```

```python
#create a graph
graph=StateGraph(LLMState)

#add nodes
graph.add_node("llm_qa",llm_Qa)

#add edges
graph.add_edge(START, "llm_qa")
graph.add_edge("llm_qa", END)

#Complie the graph into a workflow
workflows=graph.compile()


```

```python
initial_state={"question":"What is the capital of india?"}
final_state=workflows.invoke(initial_state)
print(final_state["answer"])
```
